```python
# Mrinal Pradhan
# Homework2 
# Taylor Series

import numpy as np
import matplotlib.pyplot as plt

# Taylor series for sin
def sinTaylor(x, nTerms):
    x = np.array(x).astype(float)
    approximation = np.zeros_like(x)
    term = x.copy()         
    approximation += term
    for k in range(1, nTerms):
        term *= (-x**2) / ((2*k) * (2*k + 1))
        approximation += term
    return approximation


# x from 0 to 10 step 0.1
xValues = np.arange(0, 10.0 + 0.1, 0.1)
exact = np.sin(xValues)
terms = [1, 2, 3, 4]
approximations = {}
errors = {}
for n in terms:
    approximations[n] = sinTaylor(xValues, n)
    errors[n] = np.abs(approximations[n] - exact)

# Prints out exact value, approximated values , and absolut errors for the first 4 terms.
print("x     exact      APRt1      APRt2      APRt3      APRt4      ABSerr1    ABSerr2    ABSerr3    ABSerr4")
for i in range(0, 11):
    print("{:.1f}  {: .6f}  {: .6f}  {: .6f}  {: .6f}  {: .6f}  {: .2e}  {: .2e}  {: .2e}  {: .2e}".format(
        xValues[i], exact[i], approximations[1][i],
        approximations[2][i], approximations[3][i], approxs[4][i], errors[1][i], errors[2][i], errors[3][i], errors[4][i]))

# Plot truncation error vs x for 1,2,3,4 terms
plt.figure()
for n in terms_list:
    plt.plot(xValues, errors[n], label=f"{n} term(s)")
plt.xlabel("x")
plt.ylabel("Absolute error")
plt.title("Taylor Series Truncation Error for sin(x)")
plt.show()

# Discussion
# Including more terms in the Taylor Series does reduce the truncation error especially for smaller values of x. The error becomes larger as x becomes larger because the series is centered at x=0. 
# Overall, the approximation is most accurate near zero and less accurate for larger x values.  
```
