```python
# Mrinal Pradhan
# Homework1 
# Plot Monte Carlo Pi

import math
import random
import matplotlib.pyplot as plt

# Approximates value of pi
def monteCarloPi(samples): 
    insideCircle = 0 
    for i in range(samples):
        x = random.random()
        y = random.random()
        if x**2 + y**2 <= 1:
            insideCircle += 1
    return 4 * insideCircle/samples

# Finds real pi (15 digits)
realPi = math.pi

# Sample size from 10 to 1000 in steps of 20. 
sampleSize = list(range(10, 1001, 20))
errors= []

# Find errors
for j in sampleSize:
    piEstimate = monteCarloPi(j)
    error = abs(piEstimate - realPi)
    errors.append(error)

    # Plot the error
plt.figure(figsize=(8,5))
plt.plot(sampleSize, errors, marker = "o")
plt.xlabel("Absolute Error")
plt.ylabel("Number of errors")
plt.title("Monte Carlo Pi Approximation Error")
plt.show()
```
