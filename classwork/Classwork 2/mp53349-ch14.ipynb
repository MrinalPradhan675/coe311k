```python
# Mrinal Pradhan
# Challenge 14
# Approximating Derivative of f(x)=sin(x)

import numpy as np
import matplotlib.pyplot as plt
# Setup
a, b, N = 0, 2 * np.pi, 100
x = np.linspace(a, b, N)
f = np.sin
# Step size
h = (b - a) / (N - 1)

# Derivative approximations
df_dx = (f(x + h) - f(x - h)) / (2 * h)
df_forward  = (f(x + h) - f(x)) / h
df_backward = (f(x) - f(x - h)) / h
df_true     = np.cos(x)

# 4-subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
methods = [
    ("cos(x)",df_true,"blue"),
    ("Forward Difference",df_forward,"green"),
    ("Backward Difference",df_backward,"red"),
    ("Central Difference",df_dx,"purple"),]
for ax, (title, data, color) in zip(axes.flat, methods):
    ax.plot(x, df_true, "k--", label="cos(x)", linewidth=1.5)
    ax.plot(x, data, color=color, label=title, linewidth=1.5)
    ax.set_title(title)
    ax.set_xlabel("x")
    ax.set_ylabel("f'(x)")
    ax.legend(fontsize=8)
    ax.grid(True)
plt.tight_layout()
plt.show()

# They all look similar, but the central difference is the closes to the true derivitive of cos(x). 
# The forward difference has a lower f'(x) value than the true dervivtive before the vertex and higher after it.
# Vise versa for the backward difference.
```
