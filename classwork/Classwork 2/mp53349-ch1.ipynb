```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5 * np.pi, 25)

plt.subplot(2, 2, 3)
plt.scatter(x, np.sin(x), color="blue")
plt.title("Scatter Plot", y=-.35)

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), "g")
plt.title("Cosine Line")

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x), "r")
plt.title("Sine Line")

plt.subplot(2, 2, 4)
bars = np.arange(len(x))
plt.bar(bars, np.sin(x))
plt.title("Bar Graph", y=-.35)

plt.show()
```
