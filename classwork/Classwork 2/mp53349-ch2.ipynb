```python
number = 2025
remainders = []

print("{:>12} {:>14} {:>14}".format("Division", "Quotient", "Remainder"))

while number > 0:
    quotient = number // 2
    remainder = number % 2
    print("{:>12} {:>10} {:>10}".format(str(number) + " ÷ 2", quotient, remainder))
    remainders.append(str(remainder))
    number = quotient

binary = "".join(remainders[::-1])
print("Binary representation:", binary)
```
