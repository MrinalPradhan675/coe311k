```python
# Mrinal Pradhan
# Challenge 8

import numpy as np

# function that determines if it is diagonally dominant
def isDiagonallyDominant(A):
    A = np.asarray(A, dtype=float)
    n = A.shape[0]

    for i in range(n):
        diag = abs(A[i, i])
        off_diag_sum = np.sum(np.abs(A[i])) - diag
        if diag < off_diag_sum:
            return False
    return True

# Example matrices
A1= np.array([
    [1, 3],
    [4, 1]
], dtype=float)

A2 = np.array([
    [1, 1],
    [1, 1.0001]])

# Prints whether it is diagonally dominant and displays the matrix
print("Is the matrix diagonally dominant?", isDiagonallyDominant(A1))
print(A1)
print("Is the matrix diagonally dominant?", isDiagonallyDominant(A2))
print(A2)
```
