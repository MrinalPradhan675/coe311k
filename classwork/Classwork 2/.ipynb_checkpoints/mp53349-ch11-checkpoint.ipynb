```python
# Mrinal Pradhan
# Challenge 11
# Cubic Spline vs Polynomial Fitting

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Define known data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

# Create a cubic spline interpolation of the data
cs = CubicSpline(x, y, bc_type='natural')# Generate new x values for a smooth plot
x_interp = np.linspace(0, 5, 100)
y_interp = cs(x_interp)
coeffs = np.polyfit(x, y, 5)
poly = np.poly1d(coeffs)
y_poly = poly(x_interp)

# Plot both on the same graph
plt.plot(x, y, "o", label="Data Points")
plt.plot(x_interp, y_interp, label="Cubic Spline")
plt.plot(x_interp, y_poly, label="Polynomial Fit")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Spline vs Polynomial Fit")
plt.show()

#The cubic spline fits smooth piecewise curves between each data point, while the polynomial fit uses one equation for all points and can oscillate more.
```
