```python
# Mrinal Pradhan
# Challenge 3 
# Quantifying Error

import math
import random
import matplotlib.pyplot as plt

# Approximates value of pi
def monteCarloPi(samples): 
    insideCircle = 0 
    for i in range(samples):
        x = random.random()
        y = random.random()
        if x**2 + y**2 <= 1:
            insideCircle += 1
    return 4 * insideCircle/samples

# Finds real pi (15 digits)
realPi = math.pi

# Defines number of sample and runs creates lists for errors. 
numSamples = 1000
runs = 100
absErrors = []
relErrors = []

# Runs simulation multiple times and calculates error
for j in range(runs):
    piEstimate = monteCarloPi(numSamples)
    absError = abs(realPi - piEstimate)
    relError = abs(realPi - piEstimate) / realPi
    absErrors.append(absError)
    relErrors.append(relError)
    
# Plots absolute error
plt.figure(figsize=(8,5))
plt.plot(range(1, runs + 1), absErrors, marker="o")
plt.xlabel("Run Number")
plt.ylabel("Absolute Error")
plt.title("Absolute Error of Monte Carlo Pi Approximation")
plt.show()

# Plots relative error
plt.figure(figsize=(8,5))
plt.plot(range(1, runs + 1), relErrors, marker="o")
plt.xlabel("Run Number")
plt.ylabel("Relative Error")
plt.title("Relative Error of Monte Carlo Pi Approximation")
plt.show()
```
