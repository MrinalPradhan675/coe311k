```python
# Mrinal Pradhan
# Challenge 4
# Leibniz Series for Pi

import numpy as np

# Leibniz Series
realPi = np.pi
x = 0.0

for i in range(5000):
    x += ((-1.0) ** i)/ (2.0 * i + 1.0)
    estimatedPi = x * 4.0
    
    nTerms = i + 1
    if nTerms % 100 == 0:
        relError = abs(estimatedPi - realPi) / abs(realPi)
        print("Terms: ", i + 1)
        print("Pi Estimate: ", estimatedPi)
        print("Relative Error: ", relError, "\n")   
```
